# Creating a Filter, Edge Detection

### Import resources and display image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline

# Read in the image
image = mpimg.imread('data/curved_lane.jpg')

plt.imshow(image)
```

### Convert the image to grayscale

```python
# Convert to grayscale for filtering
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

### TODO: Create a custom kernel

Below, you've been given one common type of edge detection filter: a Sobel operator.

The Sobel filter is very commonly used in edge detection and in finding patterns in intensity in an image. Applying a Sobel filter to an image is a way of **taking (an approximation) of the derivative of the image** in the x or y direction, separately. The operators look as follows.

<img src="images/sobel_ops.png" width=200 height=200>

**It's up to you to create a Sobel x operator and apply it to the given image.**

For a challenge, see if you can put the image through a series of filters: first one that blurs the image (takes an average of pixels), and then one that detects the edges.

```python
# Create a custom kernel

# 3x3 array for edge detection
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

## TODO: Create and apply a Sobel x operator
sobel_x = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image_1 = cv2.filter2D(gray, -1, sobel_y)

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image_2 = cv2.filter2D(gray, -1, sobel_x)

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)
# This time we will filter the image that filtered by sobel_y previously by sobel_x as well
filtered_image_3 = cv2.filter2D(filtered_image_1, -1, sobel_x)

# plot the images
fig=plt.figure(figsize=(18, 18))

fig.add_subplot(2, 2, 1)
plt.imshow(filtered_image_1)
plt.title('filtered by sobel_y')

fig.add_subplot(2, 2, 2)
plt.imshow(filtered_image_2)
plt.title('filtered by sobel_x')

fig.add_subplot(2, 2, 3)
plt.imshow(filtered_image_3)
plt.title('filtered by sobel_x and sobel_y')

plt.show()
```

```python
# 3x3 array for edge detection

# Read in the image
image = mpimg.imread('data/bridge_trees_example.jpg')

# my own filters
my_filter_1 = np.array([[ 0, -1, 0], 
                   [ -1, 8, -1], 
                   [ 0, -1, 0]])

my_filter_2 = np.array([[ 0, -1, 0, 4, 1], 
                        [ 0, -1, 0, 4, 1],
                        [ 0, -1, 0, 4, 1], 
                        [ 0, -1, 0, 4, 1],
                        [ 0, -1, 0, 4, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image_1 = cv2.filter2D(image, -1, my_filter_1)

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image_2 = cv2.filter2D(image, -1, my_filter_2)


# plot the images
fig=plt.figure(figsize=(18, 18))

fig.add_subplot(1, 3, 1)
plt.imshow(image)
plt.title('normal image')

fig.add_subplot(1, 3, 2)
plt.imshow(filtered_image_1)
plt.title('filtered with my_filter_1')

fig.add_subplot(1, 3, 3)
plt.imshow(filtered_image_3)
plt.title('filtered with my_filter_2')

plt.show()
```
